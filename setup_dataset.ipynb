```python
import os
import cv2

def setup_dataset_structure():
    """Create the complete dataset directory structure"""
    print("ğŸ“ Setting up dataset directory structure...")
    
    # Create main directories
    directories = [
        'dataset',
        'models'
    ]
    
    # Expression subdirectories
    expressions = ['angry', 'disgust', 'fear', 'happy', 'neutral', 'sad', 'surprise']
    
    for directory in directories:
        if not os.path.exists(directory):
            os.makedirs(directory)
            print(f"âœ… Created: {directory}/")
        else:
            print(f"ğŸ“ Already exists: {directory}/")
    
    # Create expression subdirectories
    for expression in expressions:
        expr_path = f'dataset/{expression}'
        if not os.path.exists(expr_path):
            os.makedirs(expr_path)
            print(f"âœ… Created: {expr_path}/")
    
    print("\nğŸ¯ Dataset structure created successfully!")
    return True

def check_opencv():
    """Check if OpenCV is installed and working"""
    try:
        import cv2
        print(f"âœ… OpenCV version: {cv2.__version__}")
        
        # Test camera
        cap = cv2.VideoCapture(0)
        if cap.isOpened():
            print("âœ… Camera is accessible")
            cap.release()
            return True
        else:
            print("âŒ Camera not accessible")
            return False
            
    except ImportError:
        print("âŒ OpenCV not installed")
        return False

def create_sample_dataset():
    """Create a minimal sample dataset with instructions"""
    print("\nğŸ“ Creating sample dataset structure...")
    
    expressions = ['angry', 'disgust', 'fear', 'happy', 'neutral', 'sad', 'surprise']
    
    # Create README file for dataset
    readme_content = """# Facial Expression Dataset

## Directory Structure
dataset/
â”œâ”€â”€ angry/          # ğŸ˜  Angry expression images
â”œâ”€â”€ disgust/        # ğŸ¤¢ Disgust expression images  
â”œâ”€â”€ fear/          # ğŸ˜¨ Fear expression images
â”œâ”€â”€ happy/         # ğŸ˜Š Happy/Smile expression images
â”œâ”€â”€ neutral/       # ğŸ˜ Neutral expression images
â”œâ”€â”€ sad/          # ğŸ˜¢ Sad expression images
â””â”€â”€ surprise/     # ğŸ˜² Surprise expression images

## How to Collect Data:
1. Run: python webcam_capture_fixed.py
2. Press 1-7 to select expression type
3. Make the facial expression
4. Press SPACE to capture image
5. Capture 50-100 images per expression for best results

## Minimum Requirements:
- 20+ images per expression for basic training
- 50+ images per expression for good accuracy
- 100+ images per expression for best results
"""
    
    with open('dataset/README.txt', 'w', encoding='utf-8') as f:
        f.write(readme_content)
    
    print("âœ… Sample dataset structure created!")
    print("ğŸ“– Created dataset/README.txt with instructions")

def main():
    print("ğŸ­ FACIAL EXPRESSION DATASET SETUP")
    print("=" * 50)
    
    # Setup directory structure
    setup_dataset_structure()
    
    # Check OpenCV
    if not check_opencv():
        print("\nâŒ Please install OpenCV first:")
        print("   pip install opencv-python")
        return
    
    # Create sample structure
    create_sample_dataset()
    
    print("\n" + "=" * 50)
    print("ğŸ¯ SETUP COMPLETE!")
    print("\nğŸ“‹ Next Steps:")
    print("1. ğŸ¥ Run: python webcam_capture_fixed.py")
    print("2. ğŸ“¸ Capture images for each expression")
    print("3. ğŸ§  Run: python train_expression_model_fixed.py") 
    print("4. ğŸ” Run: python main_expression_detection.py")
    print("\nğŸ’¡ Tip: Capture at least 20 images per expression to start!")

if __name__ == "__main__":
    main()
```
