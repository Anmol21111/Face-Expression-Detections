```python
import os
import sys
import subprocess

def check_requirements():
    """Check if all required packages are installed"""
    print("ğŸ” Checking requirements...")
    
    requirements = {
        'opencv-python': 'cv2',
        'tensorflow': 'tensorflow',
        'numpy': 'numpy', 
        'matplotlib': 'matplotlib'
    }
    
    missing = []
    
    for package, import_name in requirements.items():
        try:
            __import__(import_name)
            print(f"âœ… {package}")
        except ImportError:
            print(f"âŒ {package}")
            missing.append(package)
    
    return missing

def install_packages(packages):
    """Install missing packages"""
    print(f"\nğŸ“¦ Installing {len(packages)} missing packages...")
    
    for package in packages:
        try:
            subprocess.check_call([sys.executable, "-m", "pip", "install", package])
            print(f"âœ… Installed: {package}")
        except subprocess.CalledProcessError:
            print(f"âŒ Failed to install: {package}")
            return False
    
    return True

def run_setup():
    """Run the dataset setup"""
    print("\nğŸ“ Setting up dataset structure...")
    try:
        from synthetic_dataset import main as setup_main
        setup_main()
        return True
    except Exception as e:
        print(f"âŒ Setup failed: {e}")
        return False

def run_data_collection():
    """Run data collection"""
    print("\nğŸ¥ Starting data collection...")
    print("ğŸ’¡ Make different facial expressions and press SPACE to capture!")
    print("   Press 1-7 to change expression, Q to quit")
    
    try:
        from webcam_capture import DatasetCollector
        collector = DatasetCollector()
        collector.capture_dataset()
        return True
    except Exception as e:
        print(f"âŒ Data collection failed: {e}")
        return False

def run_training():
    """Run model training"""
    print("\nğŸ§  Starting model training...")
    
    try:
        from train_expression_complete import main as train_main
        train_main()
        return True
    except Exception as e:
        print(f"âŒ Training failed: {e}")
        return False

def run_detection():
    """Run real-time detection"""
    print("\nğŸ” Starting real-time expression detection...")
    
    try:
        from main_expression_detection import main as detect_main
        detect_main()
        return True
    except Exception as e:
        print(f"âŒ Detection failed: {e}")
        return False

def main():
    print("ğŸ­ COMPLETE FACIAL EXPRESSION RECOGNITION SYSTEM")
    print("=" * 60)
    
    # Check requirements
    missing_packages = check_requirements()
    
    if missing_packages:
        print(f"\nâŒ Missing {len(missing_packages)} packages")
        install = input("Do you want to install them? (y/n): ").lower().strip()
        if install == 'y':
            if not install_packages(missing_packages):
                print("âŒ Failed to install some packages. Please install manually.")
                return
        else:
            print("âŒ Please install the missing packages to continue.")
            return
    
    while True:
        print("\n" + "=" * 50)
        print("ğŸ® CHOOSE AN OPTION:")
        print("1. ğŸ—ï¸  Setup project structure (RUN THIS FIRST)")
        print("2. ğŸ¥ Collect facial expression data") 
        print("3. ğŸ§  Train expression model")
        print("4. ğŸ” Run real-time detection")
        print("5. ğŸš€ Run complete pipeline (1-4)")
        print("0. âŒ Exit")
        
        choice = input("\nEnter your choice (0-5): ").strip()
        
        if choice == '1':
            run_setup()
            
        elif choice == '2':
            run_data_collection()
            
        elif choice == '3':
            run_training()
            
        elif choice == '4':
            run_detection()
            
        elif choice == '5':
            print("\nğŸš€ RUNNING COMPLETE PIPELINE...")
            if (run_setup() and 
                run_data_collection() and 
                run_training() and 
                run_detection()):
                print("\nğŸ‰ PIPELINE COMPLETED SUCCESSFULLY!")
            else:
                print("\nâŒ Pipeline failed at some step.")
                
        elif choice == '0':
            print("ğŸ‘‹ Goodbye!")
            break
            
        else:
            print("âŒ Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
